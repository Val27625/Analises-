```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configura o estilo esttico dos grficos
sns.set_style("whitegrid")

df = pd.read_csv('/ecommerce_estatistica.csv')

# 1. Criar Histograma para 'Nota' (Avaliaes)
plt.figure(figsize=(10, 6)) # Define o tamanho da figura
sns.histplot(df['Nota'], bins=10, kde=True, color='skyblue') # Cria o histograma
plt.title('Distribuio de Notas dos Produtos', fontsize=16) # Ttulo do grfico
plt.xlabel('Notas', fontsize=12) # Rtulo do eixo X
plt.ylabel('Frequncia', fontsize=12) # Rtulo do eixo Y
plt.xticks(fontsize=10) # Tamanho da fonte dos ticks do eixo X
plt.yticks(fontsize=10) # Tamanho da fonte dos ticks do eixo Y
plt.grid(axis='y', linestyle='--', alpha=0.7) # Adiciona grade no eixo Y
plt.savefig('nota_histogram.png') # Salva o grfico como imagem PNG
plt.show() # Exibe o grfico

# Configura o estilo esttico dos grficos
sns.set_style("whitegrid")

# Criar um grfico de disperso do Nmero de Avaliaes (N_Avaliaes) vs. Nota
plt.figure(figsize=(12, 7)) # Define o tamanho da figura
sns.scatterplot(x='N_Avaliaes', y='Nota', data=df,
                hue='Nota', # Colore os pontos com base na Nota
                size='Preo', # Define o tamanho dos pontos com base no Preo
                sizes=(20, 400), # Intervalo de tamanho dos pontos
                palette='plasma', # Paleta de cores para a Nota
                alpha=0.7) # Transparncia dos pontos

plt.title('Relao entre Nmero de Avaliaes e Nota dos Produtos', fontsize=16) # Ttulo do grfico
plt.xlabel('Nmero de Avaliaes', fontsize=12) # Rtulo do eixo X
plt.ylabel('Nota', fontsize=12) # Rtulo do eixo Y
plt.xticks(fontsize=10) # Tamanho da fonte dos ticks do eixo X
plt.yticks(fontsize=10) # Tamanho da fonte dos ticks do eixo Y
plt.grid(True, linestyle='--', alpha=0.6) # Adiciona grade ao grfico
plt.legend(title='Nota', bbox_to_anchor=(1.05, 1), loc='upper left') # Adiciona legenda fora do grfico
plt.tight_layout() # Ajusta o layout para evitar sobreposio
plt.savefig('avaliacoes_nota_scatter.png') # Salva o grfico como imagem PNG
plt.show() # Exibe o grfico


# Select only numerical columns for correlation matrix
numerical_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Create a heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Mapa de Calor da Matriz de Correlao', fontsize=16)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10, rotation=0) # Ensure y-axis labels are readable
plt.tight_layout()
plt.savefig('correlation_heatmap.png')
plt.show()

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Count the occurrences of each brand and get the top 10
top_brands = df['Marca'].value_counts().head(10)

# Create a bar chart for the top 10 brands
plt.figure(figsize=(12, 7))
sns.barplot(x=top_brands.index, y=top_brands.values, palette='viridis')
plt.title('Top 10 Marcas Mais Frequentes', fontsize=16)
plt.xlabel('Marca', fontsize=12)
plt.ylabel('Nmero de Produtos', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10) # Rotate labels for better readability
plt.yticks(fontsize=10)
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.savefig('top_brands_bar_chart.png')
plt.show()

# Count the occurrences of each gender
gender_counts = df['Gnero'].value_counts()

# Create a bar chart for gender distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='plasma')
plt.title('Distribuio de Produtos por Gnero', fontsize=16)
plt.xlabel('Gnero', fontsize=12)
plt.ylabel('Nmero de Produtos', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.savefig('gender_distribution_bar_chart.png')
plt.show()

import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Count the occurrences of each season
season_counts = df['Temporada'].value_counts()

# Create a pie chart for season distribution
plt.figure(figsize=(10, 10))
plt.pie(season_counts, labels=season_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribuio de Produtos por Temporada', fontsize=16)
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.savefig('season_distribution_pie_chart.png')
plt.show()

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Create a density plot (KDE plot) for the 'Preo' column
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Preo'], fill=True, color='purple', alpha=0.6)
plt.title('Grfico de Densidade do Preo dos Produtos', fontsize=16)
plt.xlabel('Preo (R$)', fontsize=12)
plt.ylabel('Densidade', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('preco_density_plot.png')
plt.show()


# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Create a regression plot for Discount vs. Qtd_Vendidos_Cod
plt.figure(figsize=(10, 7))
sns.regplot(x='Desconto', y='Qtd_Vendidos_Cod', data=df, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title('Grfico de Regresso: Desconto vs. Quantidade Vendida', fontsize=16)
plt.xlabel('Desconto (%)', fontsize=12)
plt.ylabel('Quantidade Vendida (Codificada)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('desconto_qtd_vendidos_regression_plot.png')
plt.show()
```
